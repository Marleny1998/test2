```python
iris=datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])
iris_dataframe["target"]=iris.target
x=iris_dataframe["SL"]
print(x.mean())
y=iris_dataframe["SW"]
col=iris_dataframe["target"]
print(y.mean())
PL=iris_dataframe["PL"]*50
PW=iris_dataframe["PW"]*40
```

```python
plt.scatter(x,y,c=col)
plt.show()
```

```python
plt.scatter(x,y,c=col)
plt.xlabel("Largo del Sépalo (cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.title("Ancho del Sépalo vs Largo del Sépalo")
plt.show()
```

```python
plt.scatter(x,y,c=col)
plt.xlabel("Largo del Sépalo (cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.title("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,c=col,s=10)
plt.xlabel("Largo del Sépalo (cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.title("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,c=col,s=PL)
plt.xlabel("Largo del Sépalo (cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.title("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```
